```python
# Crypto Ape Web Scraper (5min)
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import time

# For loop that loops through all pages assuming there is less than 100
for page_number in range(1, 100):  
    url = f"https://apespace.io/api/discover?chain=bsc&page={page_number}&v=90f156c664b7e737"
    response = requests.get(url)
    data = response.json()
    
    # Loops through dictionary data to pull the 24h information on all crypto
    for i in data["data"]:
        print(f'Crypto: {i["info"]["baseName"]}\nPrice: {i["stats"]["5m"]["price"]}\nTXS: {i["stats"]["5m"]["txs"]}\nVolume: {i["stats"]["5m"]["volume"]}\n')

    # Add a delay between requests to avoid overwhelming the server
    time.sleep(1)  # Adjust the delay as needed

```
